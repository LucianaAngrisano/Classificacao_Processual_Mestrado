```python
import random
def substituir_nomes(texto, lista_nomes, nomes_ficticios):
    for nome in lista_nomes:
        if nome in texto:
            nome_ficticio = random.choice(nomes_ficticios)
            texto = texto.replace(nome, nome_ficticio)
    return texto
```

```python
# Exemplo de uso

documento_anonimizado = substituir_nomes(texto, lista_nomes, nomes_ficticios)

print(documento_anonimizado)
```

```python
import re

def substituir_cpfs_por_x(texto, substituicao="XXXX"):
    # Regex para encontrar CPFs em um texto
    padrao_cpf = r'\b\d{3}\.\d{3}\.\d{3}-\d{2}\b'
    # Substituir todos os CPFs encontrados pela string de substituição
    texto_substituido = re.sub(padrao_cpf, substituicao, texto)
    return texto_substituido

# Exemplo de uso

```

```python
texto_SEM_CPF = substituir_cpfs_por_x(documento_anonimizado)

print("Texto anonimizado:", texto_SEM_CPF)
```

```python


# Supondo que 'dados' é o seu DataFrame e que já contém as colunas 'peticaoInicial' e 'coordenacao'
# dados = pd.DataFrame(...)  # Substitua por seu carregamento de dados

# Garantir que tenhamos pelo menos 2 de cada target
min_por_grupo = 2
grupos = dados.groupby('coordenacao')

# Obter os índices de pelo menos dois membros de cada classe
indices_estratificados = grupos.apply(lambda x: x.sample(n=min_por_grupo, random_state=42)).index.get_level_values(1)

# Remover os índices já selecionados do dataset original
dados_restantes = dados.drop(indices_estratificados)

# Calcular quantos dados adicionais precisamos amostrar
num_amostras_restantes = 100 - len(indices_estratificados)

# Amostrar o restante dos dados aleatoriamente
indices_aleatorios = dados_restantes.sample(n=num_amostras_restantes, random_state=42).index

# Combinar os dois conjuntos de índices
indices_finais = indices_estratificados.union(indices_aleatorios)

# Obter as linhas finais baseadas nos índices
amostra_final = dados.loc[indices_finais]
```

```python
amostra_final
```

```python
# Supondo que 'lista_nomes' e 'nomes_ficticios' já estejam definidos
amostra_final['semnome'] = amostra_final['peticaoInicial'].apply(lambda x: substituir_nomes(x, lista_nomes, nomes_ficticios))
```

```python
amostra_final['petanonima'] = amostra_final['semnome'].apply(substituir_cpfs_por_x)
```

```python
amostra_final.iloc[0]['petanonima']
```

```python
def remover_string(texto):
    palavras_a_remover = ["sapiens.agu"]
    for palavra in palavras_a_remover:
        texto = texto.replace(palavra, "")
    return texto
```

```python
amostra_final['petanonima'] = amostra_final['petanonima'].apply(remover_string)
```
